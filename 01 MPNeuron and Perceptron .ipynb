# MP neuron class

```python
class MP_neuron:
    def __init__ (self):
        self.b = None
    def model(self,x):
        return (sum(x)>=self.b)
    def predict(self,X):
        Y=[]
        for i in X:
            result=self.model(i)
            Y.append(result)
        return np.array(Y)
    def fit(self, X,Y):
        accuracy ={}
        for b in range(X.shape[1]+1):
            self.b=b
            Y_pred = self.predict(X)
            accuracy[b] = accuracy_score(Y_pred,Y)
            
        best_b =  max(accuracy , key = accuracy.get)
        self.b=best_b
        
        print('most optimal solution of b is ' ,best_b)
        print('accuracy is ' ,accuracy[best_b])
```

```python
mp_neuron = MP_neuron()
mp_neuron.fit(X_binarised_train , Y_train)
```

Checking accuracy 

```python
Y_test_pred= mp_neuron.predict(X_binarised_test)
accuracy_Test=accuracy_score(Y_test_pred,Y_test)
print(accuracy_Test)
```

# Perceptron Class

We will introduce epochs for the convergence and observe the relation

```python
class Perceptron:
    def __init__ (self):
        self.w =None;
        self.b= None;
   
    def model(self, X):
        return 1 if(np.dot(self.w,X)>=self.b) else 0
    
    def predict(self,x):
        Y=[]
        for i in x:
            result = self.model(i)
            Y.append(result)
        return np.array(Y)
    
    def fit(self,X,Y,epochs =1 , lr =1):
        self.w = np.ones(X.shape[1])
        self.b = 0
        accuracy = {}
        max_accuracy =0;
        
        for x in range(epochs):
            for i, j  in zip(X,Y):
                y_pred =self.model(i)
                if y_pred ==1 and j==0:
                    self.w=self.w-lr*i
                    self.b=self.b-lr*1
                elif y_pred == 0  and j==1:
                    self.w=self.w+lr*i
                    self.b=self.b-lr*1
                accuracy[x] = accuracy_score(self.predict(X) ,Y)
                if(accuracy[x] > max_accuracy):
                    max_accuracy = accuracy[x]
                    chkptw = self.w
                    chkptb=self.b
                
        self.b = chkptb
        self.w =chkptw
        print(max_accuracy)
        plt.plot(accuracy.values())
        plt.show()
```
